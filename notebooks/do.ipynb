```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score, classification_report
```

```python
df_train = pd.read_csv('../data/raw/train.csv', index_col=0)
df_test = pd.read_csv('../data/raw/train.csv', index_col=0)
df_train
```

```python
numeric_cols = df_train.select_dtypes(include=[np.number]).columns
df_train[numeric_cols] = df_train[numeric_cols].fillna(df_train[numeric_cols].median())

numeric_cols = df_test.select_dtypes(include=[np.number]).columns
df_test[numeric_cols] = df_test[numeric_cols].fillna(df_test[numeric_cols].median())
```

```python
numeric_columns = df_train.select_dtypes(include=[np.number]).columns.tolist()
numeric_columns = [col for col in numeric_columns if col != 'resolution']
```

```python
X_train = df_train[numeric_columns].fillna(0)
y_train = df_train['resolution']
X_test = df_test[numeric_columns].fillna(0)
```

```python

X_train_split, X_val_split, y_train_split, y_val_split = train_test_split(
    X_train, y_train, test_size=0.2, random_state=42, stratify=y_train
)

model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train_split, y_train_split)
```

```python
val_pred = model.predict(X_val_split)
val_accuracy = f1_score(y_val_split, val_pred)

print(f"Validation accuracy: {val_accuracy:.4f}")
print("Classification report:")
print(classification_report(y_val_split, val_pred))
print()
```
